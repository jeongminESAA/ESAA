### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv('/content/gdrive/MyDrive/ESAA/data/u.user.txt', sep="|",index_col='user_id')
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
users_mean = users.groupby('occupation')['age'].mean()
users_mean
# 또는 users.groupby('occupation').age.mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
a=users.groupby(['occupation','gender']).gender.count()
b=users.groupby('occupation').gender.count()
c=(a/b)*100
c.sort_values(ascending=False).xs('M',level=1)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation')['age'].agg([max,min])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation','gender'])['age'].mean() # 칼럼명 반대면 안됨
```

### Step 8.  For each occupation present the percentage of women and men

```python
a = users.groupby('occupation')['gender'].count()
b = users.groupby('occupation')['gender'].value_counts() # value count 칼럼 안에 또 나눠진줌 별로 세줌
b/a*100
```
