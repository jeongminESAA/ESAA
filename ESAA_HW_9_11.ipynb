### DataFrame과 리스트, 딕셔너리, 넘파이 ndarray 상호 변환

####넘파이 ndarray, 리스트, 딕셔너리를 Dataframe으로 변환하기
* 1차원 형태의 리스트와 ndarray를 DataFrame으로 변환하기

```python
import numpy as np

col_name1 = ['col1']
list1 = [1,2,3]
array1 = np.array(list1)
print('array1 shape:', array1.shape)
# 리스트를 이용해 DataFrame 생성
df_list1 = pd.DataFrame(list1, columns = col_name1)
print('1차원 리스트로 만든 DataFrame:\n', df_list1)
# 넘파이 ndarray를 이용해 dataframe 생성
df_array1 = pd.DataFrame(array1, columns = col_name1)
print('1차원 ndarray로 만든 DataFrame:\n', df_array1)
```

* 2차원 형태의 리스트와 ndarray를 DataFrame으로 변환하기

```python
# 3개의 칼럼명이 필요함 (3열 리스트이므로)
col_name2 = ['col1','col2','col3']

# 2행x3열 형태의 리스트와 ndarray 생성한 뒤 이를 DataFrame으로 변환
list2 = [[1,2,3],
         [11,12,13]]
array2 = np.array(list2)
print('array2 shape:', array2.shape)
# 리스트를 이용해 DataFrame 생성
df_list2 = pd.DataFrame(list2, columns = col_name2)
print('2차원 리스트로 만든 DataFrame:\n', df_list2)
# 넘파이 ndarray를 이용해 dataframe 생성
df_array2 = pd.DataFrame(array2, columns = col_name2)
print('2차원 ndarray로 만든 DataFrame:\n', df_array2)
```

* 딕셔너리를 DataFrame으로 변환하기

### DataFrame 데이터 삭제
* drop()을 이용
> DataFrame.drop(labels = None, axis = 0, index = None, columns = None, level = None, inplace = False, errors = 'raise)

* 가장 중요한 파라미터는 labels, axis, inplace
> axis 값에 따라서 특정 로우(axis=0) 또는 칼럼(axis=1) 을 드롭  
> labels에 원하는 칼럼 명 입력하고 axis=1을 입력하면 지정된 칼럼을 드롭  
> axis=0일 경우 labels에 오는 값을 인덱스로 간주

```python
titanic_drop_df = titanic_df.drop('Age_0', axis = 1)
titanic_drop_df.head(3)
```

* Age_0 칼럼이 삭제됨

```python
titanic_df.head(3)
```

* 하지만 원본 titanic DataFrame을 보면 Age_0 칼럼이 여전히 존재함
> 이는 Inplace = False 때문임  
> 자기 자신의 DataFrame의 데이터는 삭제하지 않고 삭제된 결과만을 반환하는 것  
> True로 설정하면 자기 자신도 삭제됨
* 여러 칼럼을 삭제하고 싶으면 리스트 형태로 labels에 입력

```python
drop_result = titanic_df.drop(['Age_0','Age_by_10','Family_No'], axis=1, inplace=True)
print('inplace=True 로 drop 후 반환된 값:', drop_result)
titanic_df.head(3)
```

* inplace = True로 설정 시 반환값이 None이 되는 것 주의  
* 그렇기에 titanic_df = titanic_df.drop(inplace=True) 이렇게 다시 자신의 원래 DataFrame 객체로 할당하면 안됨
* drop_titanic 등 새로운 객체에 할당해줘야 함

### 정렬, Aggregation 함수, GroupBy 적용
#### DataFrame, Series의 정렬 - sort_values()
* sort_values()는 SQL의 order by 키워드와 매우 유사함
* 주요 파라미터는 by, ascending, inplace
> by : 해당 칼럼으로 정렬 수행  
> ascending = True : 오름차순(default)  
> inplace = False : 기존 DataFrame 유지, 정렬된 결과만 반환

```python
titanic_sorted = titanic_df.sort_values(by=['Name'])         # 이름 오름차 순으로 정렬됨
titanic_sorted.head(3)
```

```python
titanic_sorted = titanic_df.sort_values(by=['Pclass','Name'],ascending=False)                  # 내림차 순으로 정렬됨
titanic_sorted.head(3)
```

### apply lambda 식으로 데이터 가공
* 복잡한 데이터 가공이 필요할 경우 apply lambda 이용
* 함수형 프로그래밍에 사용됨

```python
def get_square(a):
    return a**2

print('3의 제곱은:', get_square(3))
```

* 원래는 get_square(a):와 같이 함수명과 입력 인자를 먼저 선언하고 결과값을 return으로 반환해야 함

```python
lambda_square = lambda x : x ** 2
print('3의 제곱은:', lambda_square(3))
```

* lambda는 이러한 것들을 한 줄로 쉽게 변환하는 식

```python
a = [1,2,3]
squares = map(lambda x : x**2, a)
list(squares)
```

* 여러 개의 값을 입력 인자로 사용할 경우에는 map()함수를 결합해서 사용함

```python
titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x : len(x))
titanic_df[['Name', 'Name_len']].head(3)
```

* Name 칼럼의 문자열 개수를 Name_len 칼럼으로 생성

```python
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x : 'Child' if x <= 15 else 'Adult')
titanic_df[['Age', 'Child_Adult']].head(8)
```

* if else를 지원함
> 주의할 점은 if절의 경우 if 식보다 반환 값을 먼저 기술해야 함  
> ex) lambda : 'Child' if x <= 15 else 'Adult
* else if는 지원하지 않음

```python
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : 'Child' if x <= 15 else ('Adult' if x <= 60 else 'Elderly'))
titanic_df['Age_cat'].value_counts()
```

* 15세 이하는 Child, 15~60세는 Adult, 61세 이상은 Elderly로 분류하는 Age_Cat 칼럼 생성

```python
# 나이에 따라 세분화된 분류를 수행하는 함수 생성
def get_category(age):
  cat = ''
  if age <= 5: cat = 'Baby'
  elif age <= 12: cat = 'Child'
  elif age <= 18: cat = 'Teenager'
  elif age <= 25: cat = 'Student'
  elif age <= 35: cat = 'Young Adult'
  elif age <= 60: cat = 'Adult'
  else : cat = 'Elderly'

  return cat

# lambda 식에 위에서 생성한 get_category() 함수를 반환값으로 지정
# get_category(X)는 입력값으로 'Age' 칼럼 값을 받아서 해당하는 cat 반환
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : get_category(x))
titanic_df[['Age', 'Age_cat']].head()
```
