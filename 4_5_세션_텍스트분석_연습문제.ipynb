```python
emma = nltk.corpus.gutenberg.raw('austen-emma.txt')
print(emma[:5000])
```

### (1) 문제 1 : emma[:5000]에서 스톱 워드 제거하기

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk import word_tokenize

emma5000 = emma[:5000]
# 단어 토큰화
emma_words = word_tokenize(emma5000)
print(len(emma_words))
print(emma_words)
```

```python
stopwords = nltk.corpus.stopwords.words('english')
filtered_words = []

for word in emma_words:
  word = word.lower()
  if word not in stopwords:
    filtered_words.append(word)

print(filtered_words)
```

### (2) 문제 2 : 스톱 워드 제거하고 남은 단어들의 원형 찾기

```python
from nltk.stem import WordNetLemmatizer
lemma = WordNetLemmatizer()

stemmed_tokens = [lemma.lemmatize(token) for token in filtered_words]
stemmed_tokens
```

```python
# 문제 2 : 문서에 대해 카운트기반 벡터화를 해보세요

#리스트 형태로 변환
emma = [emma]
words = word_tokenize(emma)

from sklearn.feature_extraction.text import CountVectorizer

# CountVectorizer 객체 생성
vectorizer = CountVectorizer()

# 텍스트 데이터를 특성 벡터로 변환
X = vectorizer.fit_transform(words)

print("특성 이름 (단어):")
print(vectorizer.get_feature_names_out())

print("\n변환된 특성 벡터:")
print(X.toarray())
```

```python
def word_tokenize(text):
    # 공백을 기준으로 단어 토큰화
    tokens = text.split()
    return tokens

# 단어 토큰화 함수 호출
tokens = word_tokenize(emma)

# 토큰 출력
print(tokens)
```

```python
# 문제 4 : 희소행렬 만들기
from sklearn.feature_extraction.text import CountVectorizer

# CountVectorizer를 사용하여 텍스트를 희소행렬로 변환
vectorizer = CountVectorizer()
sparse_matrix = vectorizer.fit_transform([emma])

# 희소행렬 출력
print("희소행렬:\n", sparse_matrix)
```

```python
from scipy.sparse import csr_matrix

# CountVectorizer를 사용하여 텍스트를 희소행렬로 변환
vectorizer = CountVectorizer()
sparse_matrix = vectorizer.fit_transform([emma])

# 희소행렬을 CSR 형식으로 변환
csr_sparse_matrix = csr_matrix(sparse_matrix.toarray())

# CSR 형식의 희소행렬 출력
print("CSR 형식의 희소행렬:\n", csr_sparse_matrix)
```
